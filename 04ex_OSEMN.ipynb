4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

```python
def to_credit_card(data):
    numbers = [int(data[i:i+3],2) for i in range(0, len(data), 6) ]
    s = ''.join(map(str, numbers))
    #remove blank
    s = s[:4] + s[5:9]+s[10:14]+s[15:]
    return int(s)

#read data
f=open("credit_card.dat","rb")
data=list(f.readlines())
#remove last 4 bits
for i in np.arange(len(data)):
    data[i]=data[i][:-5]
    
data=data[:-1]
print('credit card numbers: \n',list(map(to_credit_card,data)))
```
